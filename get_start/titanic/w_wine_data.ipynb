```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
import xgboost as xg
import requests
import io
%matplotlib inline
```

```python
data_url = "http://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
```

```python
CSV_COLUMN_NAMES = ['Class', 'Alcohol', 'MalicAcid',
                      'Ash', 'AlcalinityOfAsh', 'Magnesium', 'TotalPhenols', 'Flavanoids', 'NonflavanoidPhenols',
                      'Proanthocyanins', 'ColorIntensity', 'Hue', 'DilutedWines', 'Proline']

  # # 获取数据路径
#path = tf.keras.utils.get_file(fname=os.path.basename(data_url), origin=data_url)
#df = pd.read_csv(path)

s = requests.get(data_url).content
data = pd.read_csv(io.StringIO(s.decode('utf-8')), names=CSV_COLUMN_NAMES)
#data = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
data['Class'].value_counts().plot.pie(explode=[0,0,0],autopct='%1.1f%%', ax=ax[0],shadow=True)
ax[0].set_title('Class')
ax[0].set_ylabel('')
sns.countplot('Class',data=data,ax=ax[1])
ax[1].set_title('Class')
plt.show()
```
